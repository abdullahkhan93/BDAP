```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

# import from utilities
import stan_utility
```

```python
df = pd.read_csv("kc_house_data.csv")
random_subset = df.sample(n = 4000)
random_subset = random_subset.values
y = random_subset[:2999,2].ravel().astype(str).astype(float)
N = len(y)
x = random_subset[:2999,3:7].astype(str).astype(float)
F = 4
x_test = random_subset[3000:,3:7].astype(str).astype(float)
y_test = random_subset[3000:,2].ravel().astype(str).astype(float)
data = dict(
    N = N,
    F = F,
    y = y,
    x = x,
    x_test = x_test
)
```

```python
model = stan_utility.compile_model('model_trial.stan')
```

```python
fit = model.sampling(data = data, seed = 194838, n_jobs = -1)
samples = fit.extract(permuted = True)
print(fit)
```

```python
x_ax = np.arange(1, 51, 1)
y_ax = samples['y_predicted'].mean(axis = 0)
yreal_ax = random_subset[3000:4999, 2].astype(str).astype(float)
plt.plot(x_ax, y_ax[:50], marker='o', c = 'red')
plt.plot(x_ax, yreal_ax[:50], marker='x', c = 'green')
plt.show()
```
